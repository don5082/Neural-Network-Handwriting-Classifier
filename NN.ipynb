```python
from random import random
from random import seed
from math import exp
```

```python
def initialize_network(n_input, n_hidden, n_output):
    network = list()

    # for each input (plus one for the bias), 
    # there will be a weight connecting to each of the hidden neurons
    hidden_layer_one = [{'weights' : [random() for i in range(n_input + 1)]} for i in range(n_hidden)]
    network.append(hidden_layer_one)

    # create a second hidden layer with weights
    # and bias connected to the previous hidden layer
    hidden_layer_two = [{'weights': [random() for i in range(n_hidden + 1)]} for i in range(n_hidden)]
    network.append(hidden_layer_two)

    # Create the output layer
    output_layer = [{'weights' : [random() for i in range(n_hidden + 1)]} for i in range(n_output)]
    network.append(output_layer)

    return network
```

```python
seed(1)
network = initialize_network(2,1,2)
for layer in network:
    print(f"layer {layer.index}:{layer}")
```

```python
def activate_neuron(weights, inputs):
    activation = weights[-1] # weights[-1] = the bias
    for i in range(len(weights) - 1):
        activation += (weights[i] * inputs[i])
    return activation
```
